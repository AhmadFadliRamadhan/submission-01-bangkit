### Cleaning Data

#### Simpan dataframe dan kolom yang akan digunakan

```python
df['date'] = pd.to_datetime(df[['year', 'month', 'day']])

drop_col = ['No', 'year', 'month', 'day', 'hour', 'wd', 'WSPM', 'station']

for i in df.columns:
  if i in drop_col:
    df.drop(labels=i, axis=1, inplace=True)

df.head()
```

Kemudian bagi `df` menjadi `df_qualityair` dan `df_variable`.

```python
df_qualityair = df[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3', 'date']].copy()
df_variable = df[['TEMP', 'PRES', 'DEWP', 'RAIN', 'date']].copy()
```

Berdasarkan data-data yang diurutkan dari nilai tertinggi di atas, terlihat bahwa sepertinya angka `max` yang besar bukan merupakan **outliners**, karena angka nya menurun secara perlahan setelah diurutkan. Kecuali untuk kolom `PM2.5`, sehingga perlu di-replace dengan data terdekatnya agar tidak menjadi **outliner**.

```python
max_value = df_qualityair['PM2.5'].max()
df_qualityair['PM2.5'] = df_qualityair['PM2.5'].replace(max_value, 647.0)
df_qualityair[df_qualityair['PM2.5'] == df_qualityair['PM2.5'].max()]
```

#### Data dengan missing value

##### Dataframe df_qualityair

```python
total_data = df_qualityair.shape[0]
total_na = df_qualityair.isna().sum().sum()
total_rowna = (df_qualityair.isna().sum(axis=1) > 0).sum()

print("Total data:", total_data)
print("Total nilai yang hilang:", total_na)
print("Total baris data yang memiliki nilai yang hilang:", total_rowna, "\n")
print("Persentase baris data yang hilang: {:.2%}".format((total_rowna / total_data)))
df_qualityair.isna().sum()
```

Ayo lihat baris data yang memiliki data kosong, dimulai dari kolom `PM2.5`.

```python
df_qualityair[df_qualityair['PM2.5'].isna()]
```

Ternyata ada banyak data yang memiliki nilai kosong di ke-enam parameter kualitas udara nya. Ayo buang terlebih dahulu data dengan **missing value** di semua parameter kualitas udara nya, karena data ini tidak dapat digunakan.

```python
df_qualityair.drop(df_qualityair[df_qualityair.isna().sum(axis=1) == 6].index, inplace=True)

total_data = df_qualityair.shape[0]
total_na = df_qualityair.isna().sum().sum()
total_rowna = (df_qualityair.isna().sum(axis=1) > 0).sum()

print("Total data:", total_data)
print("Total nilai yang hilang:", total_na)
print("Total baris data yang memiliki nilai yang hilang:", total_rowna, "\n")
print("Persentase baris data yang hilang: {:.2%}".format((total_rowna / total_data)))
df_qualityair.isna().sum()
```

Terlihat bahwa masih banyak baris yang memiliki nilai kosong. Mari lihat ke kolom `CO`.

```python
df_qualityair[df_qualityair['CO'].isna()]
```

```python
print(df_qualityair['CO'].unique())
```

```python
print(df_qualityair['CO'].value_counts())
```

Karena kolom `CO` ternyata memiliki variasi nilai yang sedikit, kita bisa langsung melakukan imputasi nilai yang hilang dengan  mode nya, yaitu `200.0`.

```python
df_qualityair['CO'] = df_qualityair['CO'].fillna(df_qualityair['CO'].mode()[0])

total_data = df_qualityair.shape[0]
total_na = df_qualityair.isna().sum().sum()
total_rowna = (df_qualityair.isna().sum(axis=1) > 0).sum()

print("Total data:", total_data)
print("Total nilai yang hilang:", total_na)
print("Total baris data yang memiliki nilai yang hilang:", total_rowna, "\n")
print("Persentase baris data yang hilang: {:.2%}".format((total_rowna / total_data)))
df_qualityair.isna().sum()
```

Untuk tetap mempertahankan baris data yang memiliki nilai yang hilang, kita dapat melakukan imputasi menggunakan `intepolate()`. Alasannya karena kolom-kolom tersebut bertipe numerik, dan metode ini sangat cocok digunakan untuk menangani missing value pada data time series.

```python
df_qualityair = df_qualityair.interpolate(method='linear')

total_data = df_qualityair.shape[0]
total_na = df_qualityair.isna().sum().sum()
total_rowna = (df_qualityair.isna().sum(axis=1) > 0).sum()

print("Total data:", total_data)
print("Total nilai yang hilang:", total_na)
print("Total baris data yang memiliki nilai yang hilang:", total_rowna, "\n")
print("Persentase baris data yang hilang: {:.2%}".format((total_rowna / total_data)))
df_qualityair.isna().sum()
```

Baris dengan nilai hilang yang tersisa adalah **2 baris** dari total baris data yang ada. Ini berarti baris nilai kosong tersebut bisa saja berada di awal atau akhir tabel, yang disebabkan oleh metode `interpolate()` karena hanya dapat mengisi nilai kosong antara dua nilai yang ada. Untuk mengatasinya, dapat dengan imputasi menggunakan nilai terdekatnya.

```python
df_qualityair.head()
```

```python
df_qualityair['NO2'] = df_qualityair['NO2'].fillna(2.0)

total_data = df_qualityair.shape[0]
total_na = df_qualityair.isna().sum().sum()
total_rowna = (df_qualityair.isna().sum(axis=1) > 0).sum()

print("Total data:", total_data)
print("Total nilai yang hilang:", total_na)
print("Total baris data yang memiliki nilai yang hilang:", total_rowna, "\n")
print("Persentase baris data yang hilang: {:.2%}".format((total_rowna / total_data)))
df_qualityair.isna().sum()
```

```python
df_qualityair.head()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
scatter_data = df_qualityair[['PM2.5', 'PM10', 'SO2', 'NO2', 'CO', 'O3']]
scatter_matrix = pd.plotting.scatter_matrix(scatter_data, figsize=(12, 12), diagonal='kde')

for i in range(len(scatter_data.columns)):
    for j in range(len(scatter_data.columns)):
        ax = scatter_matrix[i, j]
        ax.xaxis.label.set_rotation(45)
        ax.yaxis.label.set_rotation(0)
        ax.yaxis.label.set_ha('right')
        ax.xaxis.label.set_size(10)
        ax.yaxis.label.set_size(10)

plt.suptitle('Scatter Plot Matrix Parameter Kualitas Udara di Stasiun Pemantauan Dingling ', fontsize=16)
plt.tight_layout()
plt.show()
```

Berdasarkan visualisasi scatter matriks di atas, terdapat beberapa pernyataan yang didapatkan:
- `PM2.5` dan `PM10` terlihat dengan jelas membentuk sebuah garis mengarah ke kanan atas, menunjukkan **korelasi positif yang kuat**.
- `CO` dan `NO2` membentuk persebaran mengarah ke kanan atas, persebarannya agak tersebar, menunjukkan **korelasi positif yang sedang**.
- `CO` - `PM2.5`, `NO2` - `PM2.5`, `CO` - `PM10`, dan `NO2` - `PM10` membentuk persebaran mengarah ke kanan atas walau tidak terlalu condong, memiliki persebaran yang luas, menunjukkan **korelasi positif yang lemah**.
- Sisa pasangan yang tidak disebutkan memiliki persebaran yang acak.

### Pertanyaan 2:

```python
daily_mean = df_variable.groupby(df_variable['date'].dt.date).mean()

plt.figure(figsize=(18, 6))
for column in daily_mean.columns:
    if column != 'date':
        plt.plot(daily_mean.index, daily_mean[column], label=column)

plt.title('Grafik Rata-Rata Nilai Tiap Kolom per Hari di Stasiun Pemantauan Dingling')
plt.xlabel('Tanggal')
plt.ylabel('Nilai Rata-Rata')
plt.legend()
plt.show()
```

Berdasarkan visualisasi di atas, diketahui bahwa:
- `TEMP`, `PRES`, dan `DEWP` memiliki pola musiman, di mana `TEMP` dan `DEWP` memiliki pola garis yang cenderung mirip.
- `PRES` terhadap `TEMP` dan `DEWP` memiliki pola garis yang berlawanan.
- `RAIN` memiliki pola garis lurus yang cenderung konstan pada nilai nol.
